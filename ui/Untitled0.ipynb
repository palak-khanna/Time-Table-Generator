```python
import tkinter as tk
from tkinter import messagebox
import sys
import os
import random
# Correct way to get the path of the script directory
sys.path.append(os.path.abspath(os.path.join(os.path.dirname(__file__), '..', 'script')))

# Now import the genetic algorithm and display schedule modules
from timetable_generator.scripts.genetic_algorithm import run_genetic_algorithm  # This should work now
from timetable_generator.ui.display_schedule import display_schedule

# Function to handle running the genetic algorithm and displaying the timetable
def generate_timetable():
    try:
        # Call genetic algorithm (Modify as per your function)
        best_schedule = run_genetic_algorithm()
        # Display the best schedule (modify this based on your UI needs)
        display_schedule(best_schedule)  # Assuming it will print or render the timetable
    except Exception as e:
        messagebox.showerror("Error", f"An error occurred: {e}")

# Create the main UI window
root = tk.Tk()
root.title("Timetable Generator")
root.geometry("500x400")

# Button to generate timetable
generate_button = tk.Button(root, text="Generate Timetable", command=generate_timetable)
generate_button.pack(pady=20)

# Add a Text widget for the timetable display
timetable_display = tk.Text(root, height=10, width=50)
timetable_display.pack(pady=20)

# Run the UI
root.mainloop()
```
